```python
# importing numpy, matplotlib and skfuzzy libraries
import skfuzzy as fuzz
import numpy as np
from skfuzzy import control as ctrl

print("Libraries have been imported.")
```

```python
# antecedets and consequents (inputs and outputs)
openness = ctrl.Antecedent(np.arange(0, 11), 'openness')
agreeableness = ctrl.Antecedent(np.arange(0, 11), 'agreeableness')
extraversion = ctrl.Antecedent(np.arange(0, 11), 'extraversion')
neuroticism = ctrl.Antecedent(np.arange(0, 11), 'neuroticism')
contentiousness = ctrl.Antecedent(np.arange(0, 11), 'contentiousness')
print("The input variables are as follows:")
print("1. OPENNESS : accessibility")
print("2. AGREEABLENESS : willing to do or allow something")
print("3. EXTRAVERSION : outgoing and social")
print("4. NEUROTICISM : anxiety, depression, self doubt and negatives")
print("5. CONTENTIOUSNESS : careful, efficient and organized")
print(" ")

average = ctrl.Consequent(np.arange(0, 11), 'average')
reserved = ctrl.Consequent(np.arange(0, 11), 'reserved')
role_models = ctrl.Consequent(np.arange(0, 11), 'role_models')
self_centered = ctrl.Consequent(np.arange(0, 11), 'self_centered')
print("The output variables are:")
print("1. AVERAGE")
print("2. RESERVED")
print("3. ROLE MODELS")
print("4. SELF CENTERED")
print(" ")
print("Based on the inputs given by the user, the rules will be applied and the best suited output will be generated according to that")
```

```python
# auto membership function(3,5 or 7)
openness.automf(3)
agreeableness.automf(3)
extraversion.automf(3)
neuroticism.automf(3)
contentiousness.automf(3)
print("Auto membership functions have been generated.")
```

```python
# custom membership function
average['low'] = fuzz.trimf(average.universe, [0, 0, 5])
average['medium'] = fuzz.trimf(average.universe, [0, 5, 10])
average['high'] = fuzz.trimf(average.universe, [5, 10, 10])

reserved['low'] = fuzz.trimf(reserved.universe, [0, 0, 5])
reserved['medium'] = fuzz.trimf(reserved.universe, [0, 5, 10])
reserved['high'] = fuzz.trimf(reserved.universe, [5, 10, 10])

role_models['low'] = fuzz.trimf(role_models.universe, [0, 0, 5])
role_models['medium'] = fuzz.trimf(role_models.universe, [0, 5, 10])
role_models['high'] = fuzz.trimf(role_models.universe, [5, 10, 10])

self_centered['low'] = fuzz.trimf(self_centered.universe, [0, 0, 5])
self_centered['medium'] = fuzz.trimf(self_centered.universe, [0, 5, 10])
self_centered['high'] = fuzz.trimf(self_centered.universe, [5, 10, 10])
print("Each of the input variable will be associated with the three terms, viz.,'good', 'poor' and 'average'.")
print("Each of the output variable will be associated with three terms, viz., 'low', 'medium' and 'high'.")
```

```python
# seeing these
def ViewMembership():
    import matplotlib.pyplot
    %matplotlib inline
    matplotlib.use('TkAgg')
    print("The graphs of all have been showed.")
    openness.view()
    agreeableness.view()
    extraversion.view()
    neuroticism.view()
    contentiousness.view()
    average.view()
    reserved.view()
    role_models.view()
    self_centered.view()

choice = input("Do you want to see Membership Functions?(Y/N) = ")
if(choice == "y" or choice == "Y"):
    ViewMembership()
```

```python
# rules defined (relation between inputs and outputs)
rule1 = ctrl.Rule(openness['good'] | agreeableness['good'] | extraversion['poor'] | neuroticism['poor'] | contentiousness['good'] , average['low'])
rule2 = ctrl.Rule(openness['average'] | agreeableness['average'] | extraversion['good'] | neuroticism['good'] | contentiousness['average'] , average['medium'])
rule3 = ctrl.Rule(openness['poor'] | agreeableness['poor'] | extraversion['good'] | neuroticism['good'] | contentiousness['poor'] , average['high'])

rule4 = ctrl.Rule(openness['good'] | agreeableness['poor'] | extraversion['good'] | neuroticism['good'] | contentiousness['poor'] , reserved['low'])
rule5 = ctrl.Rule(openness['average'] | agreeableness['good'] | extraversion['average'] | neuroticism['average'] | contentiousness['good'] , reserved['medium'])
rule6 = ctrl.Rule(openness['poor'] | agreeableness['good'] | extraversion['poor'] | neuroticism['poor'] | contentiousness['good'] , reserved['high'])

rule7 = ctrl.Rule(openness['poor'] | agreeableness['poor'] | extraversion['poor'] | neuroticism['good'] | contentiousness['poor'] , role_models['low'])
rule8 = ctrl.Rule(openness['good'] | agreeableness['good'] | extraversion['good'] | neuroticism['average'] | contentiousness['good'] , role_models['medium'])
rule9 = ctrl.Rule(openness['good'] | agreeableness['good'] | extraversion['good'] | neuroticism['poor'] | contentiousness['good'] , role_models['high'])

rule10 = ctrl.Rule(openness['good'] | agreeableness['good'] | extraversion['poor'] | neuroticism['poor'] | contentiousness['good'] , self_centered['low'])
rule11 = ctrl.Rule(openness['average'] | agreeableness['average'] | extraversion['good'] | neuroticism['good'] | contentiousness['average'] , self_centered['medium'])
rule12 = ctrl.Rule(openness['poor'] | agreeableness['poor'] | extraversion['good'] | neuroticism['good'] | contentiousness['poor'] , self_centered['high'])

print("There are 12 rules generated to get the outcome relevant to each input.")
```

```python
# viewing rules
def viewRule():    
    import matplotlib.pyplot
    %matplotlib inline
    print("These rules are:")
    rule1.view()
    rule2.view()
    rule3.view()
    rule4.view()
    rule5.view()
    rule6.view()
    rule7.view()
    rule8.view()
    rule9.view()
    rule10.view()
    rule11.view()
    rule12.view()

choice = input("Do you want to see rules?(Y/N) = ")
if(choice == "y" or choice == "Y"):
    viewRule()
```

```python
# creating control system
average_c = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11, rule12])
reserved_c = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11, rule12])
role_models_c = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11, rule12])
self_centered_c = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11, rule12])
print("Control system has been created for all outputs.")

# simulating control system
average_ = ctrl.ControlSystemSimulation(average_c)
reserved_ = ctrl.ControlSystemSimulation(reserved_c)
role_models_ = ctrl.ControlSystemSimulation(role_models_c)
self_centered_ = ctrl.ControlSystemSimulation(self_centered_c)
print("Simulation of them has been done.")
```

```python
# passing inputs
average_.input['openness'] = float(input("Openness for AVERAGE = "))
average_.input['agreeableness'] = float(input("Agreeableness for AVERAGE = "))
average_.input['extraversion'] = float(input("Extraversion for AVERAGE = "))
average_.input['neuroticism'] = float(input("Neuroticism for AVERAGE = "))
average_.input['contentiousness'] = float(input("Contentiousness for AVERAGE = "))

reserved_.input['openness'] = float(input("Openness for RESERVED = "))
reserved_.input['agreeableness'] = float(input("Agreeableness for RESERVED = "))
reserved_.input['extraversion'] = float(input("Extraversion for RESERVED = "))
reserved_.input['neuroticism'] = float(input("Neuroticism for RESERVED = "))
reserved_.input['contentiousness'] = float(input("Contentiousness for RESERVED = "))

role_models_.input['openness'] = float(input("Openness for ROLE MODEL = "))
role_models_.input['agreeableness'] = float(input("Agreeableness for ROLE MODEL = "))
role_models_.input['extraversion'] = float(input("Extraversion for ROLE MODEL = "))
role_models_.input['neuroticism'] = float(input("Neuroticism for ROLE MODEL = "))
role_models_.input['contentiousness'] = float(input("Contentiousness for ROLE MODEL = "))

self_centered_.input['openness'] = float(input("Openness for SELF CENTERED = "))
self_centered_.input['agreeableness'] = float(input("Agreeableness for SELF CENTERED = "))
self_centered_.input['extraversion'] = float(input("Extraversion for SELF CENTERED = "))
self_centered_.input['neuroticism'] = float(input("Neuroticism for SELF CENTERED = "))
self_centered_.input['contentiousness'] = float(input("Contentiousness for SELF CENTERED = "))
print("Individual input has been taken.")
```

```python
# computing them
average_.compute()
reserved_.compute()
role_models_.compute()
self_centered_.compute()
print("All 4 has been computed.")
```

```python
# printing the final output values
print(average_.output['average'])
print(reserved_.output['reserved'])
print(role_models_.output['role_models'])
print(self_centered_.output['self_centered'])
```
